```python
from sklearn.impute import SimpleImputer
import numpy as np

imputer = SimpleImputer(strategy="median")
df[["Titer","Purity","SEC %Monomer","SMAC","HIC"]] = imputer.fit_transform(df[["Titer","Purity","SEC %Monomer","SMAC","HIC"]])
```

```python
df["agg_label"] = (df["SEC %Monomer"] < 90).astype(int)
```

```python
train_df = df[df["random_fold"] < 4]  # training folds
test_df  = df[df["random_fold"] == 4] # holdout test
```

```python
from collections import Counter
import pandas as pd

def aa_composition(seq):
    counts = Counter(seq)
    total = len(seq)
    return {aa: counts.get(aa,0)/total for aa in "ACDEFGHIKLMNPQRSTVWY"}

# Example: heavy chain features
hc_features = df["hc_protein_sequence"].apply(aa_composition)
hc_features_df = pd.DataFrame(list(hc_features)).add_prefix("HC_")

# Light chain features
lc_features = df["lc_protein_sequence"].apply(aa_composition)
lc_features_df = pd.DataFrame(list(lc_features)).add_prefix("LC_")

# Merge into one features_df
features_df = pd.concat([hc_features_df, lc_features_df], axis=1)
```

```python
# Label: 1 if SEC % Monomer < 90 (aggregation-prone), else 0
df["agg_label"] = (df["SEC %Monomer"] < 90).astype(int)

# Final dataset
X = features_df
y = df["agg_label"]
```

```python
X
```

```python
y
```

```python
train_idx = df[df["random_fold"] < 4].index
test_idx  = df[df["random_fold"] == 4].index

X_train, y_train = X.loc[train_idx], y.loc[train_idx]
X_test,  y_test  = X.loc[test_idx], y.loc[test_idx]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

clf = RandomForestClassifier(n_estimators=200, random_state=42)
clf.fit(X_train, y_train)

y_pred = clf.predict_proba(X_test)[:,1]
print("Test ROC-AUC:", roc_auc_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import numpy as np

importances = clf.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10,5))
plt.bar(range(len(importances)), importances[indices])
plt.xticks(range(len(importances)), X.columns[indices], rotation=90)
plt.title("Feature Importance (AA composition)")
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, (y_pred > 0.5).astype(int)))
print(confusion_matrix(y_test, (y_pred > 0.5).astype(int)))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# Refit the model
clf = RandomForestClassifier(n_estimators=200, class_weight="balanced", random_state=42)
clf.fit(X_train, y_train)

# Get probabilities for class 1 (aggregation-prone)
y_prob = clf.predict_proba(X_test)[:, 1]

# Compute precision-recall curve
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

# Plot
plt.plot(recall, precision, marker='.')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.show()
```

```python
from sklearn.metrics import f1_score
import numpy as np

best_f1, best_thresh = 0, 0
for t in thresholds:
    y_pred_thresh = (y_prob >= t).astype(int)
    f1 = f1_score(y_test, y_pred_thresh)
    if f1 > best_f1:
        best_f1, best_thresh = f1, t

print("Best Threshold:", best_thresh)
print("Best F1 Score:", best_f1)

# Recalculate metrics with best threshold
y_pred_best = (y_prob >= best_thresh).astype(int)
print(classification_report(y_test, y_pred_best))
print(confusion_matrix(y_test, y_pred_best))
```
